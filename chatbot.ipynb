```python
while True:
    user_question = input("Ask a question: ")

    found = False

    for i in range(len(data)):
        if user_question.lower() == data["Question"][i].lower():
            print("Bot:", data["Answer"][i])
            found = True
            break

    if not found:
        print("Bot: Sorry, I don't know the answer.")
```
